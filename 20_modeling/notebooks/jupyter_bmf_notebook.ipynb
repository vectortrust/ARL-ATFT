```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Function, diff

x, y, t = symbols('x y t')
P = Function('P')(x, y, t)
print("∂P/∂t =", diff(P, t))

A = np.array([[2, 1], [1, 3]])
print("A⁻¹ =", np.linalg.inv(A))

plt.plot([0, 1, 2], [1, 4, 9])
plt.title("Matplotlib Test")
plt.show()
```
